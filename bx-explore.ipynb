```r
library(dplyr)
library(ggplot2)
```

```r
bx.ratings = read.csv("build/ratings.csv", col.names=c("User", "Item", "Rating"), header = FALSE)
```

## Summarize Items

How are item rating counts distributed?

```r
item.summary = bx.ratings %>%
    group_by(Item) %>%
    summarize(NRatings = n())
```

```r
options(repr.plot.height=3)
ggplot(item.summary) +
    aes(x=NRatings) +
    geom_histogram(binwidth=0.1) +
    scale_x_log10()
```

```r
item.summary %>% group_by(NRatings) %>% summarize(NItems=n()) %>% arrange(-NItems) %>% head()
```
